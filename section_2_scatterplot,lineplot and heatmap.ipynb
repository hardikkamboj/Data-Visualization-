```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg_df = sns.load_dataset('mpg')
print(mpg_df.shape)
mpg_df.head()
```

## Scatteplot

```python
#scatterplot
ax = sns.scatterplot(x = 'weight',y = 'mpg',data = mpg_df)
```

```python
#include ticks on the axis
sns.set(style = 'ticks')
##hexbin plot
sns.jointplot(x = mpg_df['weight'],y = mpg_df['mpg'],kind = 'hex')
```

## Contour Plot

```python

#contour plots
sns.kdeplot(mpg_df['weight'],mpg_df['mpg'],shade = True)
```

```python
sns.__version__
```

## Lineplot

```python
#through scatter 
sns.scatterplot(x = 'model_year',y = 'mpg',data = mpg_df)
```

```python
#through line
sns.lineplot(mpg_df['model_year'],mpg_df['mpg'])
```
