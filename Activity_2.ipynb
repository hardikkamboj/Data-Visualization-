```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'E:\programming\dataset\visualization_datsets\athlete_events.csv')
print(data.shape)
data.head()
```

```python
#filtering datasets, selecting only the rows which won any medal
data = data.dropna(subset=['Medal'])
```

```python
data = data[data['Year'] == 2016]
```

```python
top_sport = ['Rowing','Swimming','Athletics', 'Football','Hockey' ]
```

```python
data = data[data['Sport'].isin(top_sport)]
```

```python
print(data.shape)
data.head()
```

## Create a plot using an appropriate visualization technique that best presents the global pattern of the height and weight features of the 2016 medal winners of the five sports.



```python
sns.scatterplot(data['Height'],data['Weight'])
```

```python
sns.jointplot(data['Height'],data['Weight'],kind = 'hex')
```
