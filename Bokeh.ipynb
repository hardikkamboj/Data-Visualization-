## Line plots

```python
from bokeh.models import HoverTool

#data
x = np.arange(20)
y = np.random.rand(20)

#line plot
line_plot = figure(plot_width = 500,plot_height = 325, title = 'Line plot')
line_plot.line(x,y,line_width = 2)

#add hover tool
line_plot.add_tools(HoverTool())

show(line_plot)
```

### LIne plot with several lines

```python
x = np.arange(10)
y1 = np.random.rand(10)
y2 = np.random.rand(10)
y3 = np.random.rand(10)

multi_line_plot = figure(plot_width = 500,plot_height = 325, title = 'multi line plot',toolbar_location = 'below')
multi_line_plot.line(x,y1,color = 'red', line_width = 3)
multi_line_plot.line(x,y2,color = 'blue',line_width = 3)
multi_line_plot.line(x,y3,color = 'green',line_width = 3)

show(multi_line_plot)
```
